## **Project Summary**

The "Mental Health in Tech Workplace" project leverages a dataset from a 2014 survey that aimed to assess mental health awareness, treatment, stigma, and workplace support systems within the technology industry. As the global tech workforce expands, mental health concerns are becoming more prominent due to stress, long hours, job insecurity, and lack of support systems. This analysis aims to uncover key patterns and insights related to how mental health is perceived, addressed, and discussed in tech workplaces.

The dataset contains responses from over 1,200 participants and includes various features such as age, gender, country, employment type, company size, and multiple survey questions that explore mental health support, perceived consequences of disclosure, benefits availability, and openness toward mental health conversations.

Key objectives of this project include:

Understanding the demographic distribution of respondents

Investigating the relationship between workplace factors (benefits, support systems, anonymity) and mental health treatment

Exploring cultural or geographic differences in attitudes toward mental health

Analyzing whether company size, employment type, and gender influence mental health awareness or stigma

Using Python libraries such as Pandas, Seaborn, Matplotlib, and Plotly, we performed structured univariate, bivariate, and multivariate visualizations to reveal meaningful insights. We identified factors like family history, work interference, availability of mental health resources, and perceived consequences as influential in whether individuals seek treatment. The project also raises awareness about how mental health conversations can be normalized in corporate settings.

The findings aim to drive actionable suggestions for HR departments, tech organizations, and policy advocates to create more inclusive and supportive environments for their workforce.

## **Define Your Business Objective**

Despite increasing awareness of mental health in modern workplaces, many employees in the tech industry still face barriers to accessing support, fear of stigma, and uncertainty about company policies. This project seeks to identify the key barriers and enablers related to mental health support in tech workplaces based on demographic and organizational attributes.

```python
# Import Required Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
from google.colab import files
uploaded = files.upload()
```

```python
import io
df = pd.read_csv(io.BytesIO(uploaded['survey.csv']))
print(df.shape)
df.head()
```

```python
# Dataset Info
df.info()
```

```python
# Describe numerical fields
df.describe(include='all')
```

```python
# Missing values check
df.isnull().sum()
```

```python
# Data Cleaning
df = df[(df['Age'] > 18) & (df['Age'] < 65)]
```

```python
# Normalize Gender
def clean_gender(gender):
    gender = str(gender).lower()
    if "female" in gender or "f" in gender:
        return "Female"
    elif "male" in gender or "m" in gender:
        return "Male"
    elif "trans" in gender:
        return "Trans"
    else:
        return "Other"

df['Gender'] = df['Gender'].apply(clean_gender)
```

```python
# Handle Missing Values
# Fill missing values in categorical features with mode
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df.isnull().sum()
```

```python
# Age Distribution
sns.histplot(df['Age'], bins=30, kde=True, color='teal')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

**Why this chart?** To understand the age spread of tech employees responding to the survey.

**Insight:** Most respondents are between 25 and 35 years old, indicating this age group may experience the highest mental health concerns or is more responsive to such surveys.

**Business Impact**:Target mental health initiatives to younger/mid-career professionals, who form the bulk of the tech workforce.

```python
# Country Distribution (Top 10)
top_countries = df['Country'].value_counts().nlargest(10)
sns.barplot(x=top_countries.index, y=top_countries.values, hue=top_countries.index, palette="Set2", legend=False)
plt.title("Top 10 Countries of Respondents")
plt.xticks(rotation=45)
plt.show()
```

**Why this chart?** To understand geographic representation.

**Insight**: The majority are from the United States, followed by Canada and UK.

**Business Impact**: Region-specific policies or awareness campaigns (e.g., for the U.S.) may have more significant reach.

```python
# Family History
sns.countplot(data=df, x='family_history', hue='family_history', palette='muted', legend=False)
plt.title("Family History of Mental Illness")
plt.show()
```

**Why this chart? **To evaluate hereditary or background risk factors.

**Insight:** A substantial portion (~45%) reported having a family history of mental illness.

**Business Impact:** This suggests an opportunity for proactive screening and support for at-risk employees.



```python
# Sought Treatment
sns.countplot(data=df, x='treatment', hue='treatment', palette='Set1', legend=False)
plt.title("Have Respondents Sought Treatment?")
plt.show()
```

**Insight:** Those with higher interference are much more likely to seek treatment.

```python
# Create a copy for encoding
df_encoded = df.copy()
# Map binary and ordinal answers to numerical values
binary_map = {
    'Yes': 1, 'No': 0,
    "Don't know": np.nan, 'Maybe': 0.5,
    'Never': 0, 'Rarely': 1, 'Sometimes': 2, 'Often': 3,
    'Somewhat easy': 2, 'Somewhat difficult': 1, 'Very difficult': 0, 'Very easy': 3
}

# Apply mapping to relevant columns
cols_to_map = ['treatment', 'family_history', 'remote_work', 'benefits', 'care_options',
               'wellness_program', 'seek_help', 'anonymity', 'mental_health_consequence',
               'phys_health_consequence', 'mental_vs_physical', 'obs_consequence',
               'work_interfere', 'leave']

for col in cols_to_map:
    df_encoded[col] = df_encoded[col].map(binary_map)

# Drop rows with missing values
correlation_data = df_encoded[cols_to_map + ['Age']].dropna()

# Plot the correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_data.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap: Mental Health Related Variables")
plt.tight_layout()
plt.show()
```

**Insight:**
The heatmap shows that seeking treatment is strongly correlated with factors like family history, work interference, and awareness of mental health resources. Age has minimal influence.

**Business Impact:**
Companies should focus on offering mental health benefits, reducing stigma, and supporting employees with family history or high stress levels to improve mental well-being and productivity.

```python
# Sample pairplot on select features
selected_cols = ['treatment', 'family_history', 'benefits', 'work_interfere', 'leave']

# Ensure numeric values are used
pairplot_data = df_encoded[selected_cols].dropna()

sns.pairplot(pairplot_data)
plt.suptitle("Pairplot: Mental Health Related Variables", y=1.02)
plt.show()
```

**Insight:**
The pairplot reveals visible clustering and positive patterns between:

* family_history, work_interfere, and treatment

* Employees who report family history or work interference are more likely to seek treatment

benefits and leave policies show slight grouping, indicating they influence support perception

**Business Impact:**
Organizations should prioritize:

* Early mental health interventions for employees
reporting work stress

* Ensuring benefits and leave policies are not only provided but clearly communicated

* Recognizing high-risk employees (e.g., with family history) for proactive support

#**Solution to Business Objective**

To help tech organizations improve employee mental health outcomes:

1. Enhance Mental Health Support Systems

 * Provide comprehensive mental health benefits, including therapy access, wellness programs, and self-help resources.

2. Promote Mental Health Awareness

 * Regular workshops, campaigns, and mental health days can reduce stigma and encourage early help-seeking.

3. Train Supervisors and Managers

 * Equip team leads with basic mental health awareness to identify struggling employees and respond supportively.

4. Protect Anonymity and Confidentiality

 * Ensure policies clearly communicate that seeking help has no negative consequences on employment or growth.

5. Target High-Risk Groups

 * Employees with family history, work interference, or low perceived support should be proactively engaged through surveys, wellness check-ins, or 1-on-1 sessions.

6. Promote Work-Life Balance

  * Prevent work-related stress with flexible hours, remote work options, and manageable workloads.

These initiatives can significantly increase employee satisfaction, productivity, and retention while fostering a healthier workplace culture.
